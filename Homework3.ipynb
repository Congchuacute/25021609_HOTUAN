```python
#b16
x = float(input())

duoi = int(x)

if x == duoi:
    tren = duoi
else:
    tren = duoi + 1

if x - duoi >= 0.5:
    gan = duoi + 1
else:
    gan = duoi

print(tren, duoi, gan)
```
