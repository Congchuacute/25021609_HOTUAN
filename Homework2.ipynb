```python
#b29
def lt(n):
    s=1
    for i in range(1,n+1):
        s*=i
    return s
def gt(k,n):
    return(lt(n)//(lt(k)*(lt(n-k))))
n = int(input())
for i in range(0,n+1):
    print(' '*(n - i) , end='')
    for j in range(0,i+1):
        print(gt(j,i), end = ' ')
    print("\n")
    

```

```python
#b30
def hv(a):
    return a*a

def ht(r):
    return r*r*3.14

def tg(a,h):
    return a*h/2

def cn(d,r):
    return d*r

n = int(input())
if(n==1):
    a=int(input())
    print(hv(a))
if(n==2):
    r=int(input())
    print(ht(r))
if(n==3):
    a,h =map(int,input().split())
    print(tg(a,h))
if(n==4):
    d,r =map(int,input().split())
    print(cn(d,r))

```

```python
#b31

def fibo(n):
    if(n==1):
        return 1
    if(n==2):
        return 1
    s=fibo(n-1) + fibo(n-2)
    return s
n = int(input())
print(fibo(n))
```
