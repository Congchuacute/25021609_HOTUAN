```python
#b3
a,b,c = map(int, input().split())
if( a + b <= c or b + c <= a or c + a <= b):
    print("Khong phai la tam giac")
else:
    if( a == b == c):
        print("Tam giac deu")
    elif( a*a==b*b+c*c or a*a+b*b==c*c or a*a+c*c == b*b):
        print("Tam giac vuong")
    elif( a==b!=c or  a!=b==c or  a==c!=b ):
        print("Tam giac can")
    else:
        print("Tam giac thuong")
        
    
```
