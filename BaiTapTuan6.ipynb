```python
#W6A12
n = list(map( int , input().split()))
d = {}
ans = 100000
ln = -100000
for i in n:
    d[i] = d.get(i,0) + 1
    if d[i] == ln:
        if i < ans:
            ln = d[i]
            ans = i
    elif d[i] > ln:
        ln = d[i]
        ans = i
        
print(ans)
    
```
