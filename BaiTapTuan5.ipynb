```python
#W5A2
def swap( a, b ):
    a , b = b , a
    return a , b
a , b = map( int , input().split())
print(type(swap(a,b)))
```

```python
#W5A3
def snt(n):
    c = [True]*10**7
    c[0] = False
    c[1] = False
    for i in range( 2 , int((10**7)**0.5 + 1) ):
        if c[i] == True:
            for j in range( i*i , 10**7 , i):
                c[j] = False
    return c[n]
n = int(input())
snt(n)
```

```python
#W5A4
def tonguoc(n):
    s = 1
    for i in range( 2 , int(n**0.5) + 1 ):
        if n % i == 0:
            j = n / i
            if i != j:
                s = s + i + j
            else: 
                s += i
    return s == n
n = int(input())
tonguoc(n)
```

```python
#W5A6
def gt(n):
    s = 1
    for i in range( 1 , n+1 ):
        s *= i
    return s
n = int(input())
gt(n)
        
```

```python
#W5A7
def mt( a , b , k):
    if k == '+':
        return a + b
    if k == '-':
        return a - b
    if k == '*':
        return a * b
    if k == '/':
        return round(a/b , 2)
a , b = map( int , input().split())
k = input()
mt( a , b , k)

        
```

```python
#W5A8
def np(n):
    d = 0
    while n > 0:
        c = n % 2
        if c == 1:
             d += 1
        n //= 2
    return d
    
k = int(input())
m = int(input())
x = k ^ m
print(np(x))


```

```python
#W5A9
def cso(n):
    s = 0
    while n > 0:
        c = int(n % 10)
        s += c
        n = n//10
    return s
n = int(input())
cso(n)
```

```python
#W5A10
def dcau(a,b):
    mp = {}
    used = set()
    if len(a) != len(b):
        return False
    for i in range(len(a)):
        if a[i] in mp:
            if mp[a[i]] != b[i]:
                return False
        else:
            if b[i] in used:
                return False
            mp[a[i]] = b[i]
            used.add(b[i])
    return True
a = input()
b = input()
dcau(a,b)
        
        
```
