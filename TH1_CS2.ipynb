```python
#W1A1
a = 7      # Gán giá trị 7 cho biến a
b = 5      # Gán giá trị 5 cho biến b
c = a - b  # Thực hiện phép trừ: 7 - 5 = 2, gán kết quả cho biến c
print(c)   # In giá trị của biến c ra màn hình, kết quả là 2
```

```python
#W1A2
city = "Hà Nội"                     # Gán chuỗi "Hà Nội" cho biến city
year = 2025                         # Gán số nguyên 2025 cho biến year
print("Thành phố: ", city, " - Năm: ", year)  
# Hàm print sẽ in lần lượt các đối số:
# "Thành phố: " + city + " - Năm: " + year
# Khi dùng dấu phẩy trong print, Python tự động thêm khoảng trắng giữa các thành phần
# Kết quả in ra: Thành phố:  Hà Nội  - Năm:  2025
```

```python
#W1A3
n = 4               # Gán giá trị 4 cho biến n (số lượng phần tử cần cộng dồn)
t = 0               # Gán giá trị 0 cho biến t (biến tích lũy tổng)

for i in range(1, n + 1):   # Vòng lặp chạy i từ 1 đến n (bao gồm 1, 2, 3, 4)
    t += i                  # Cộng dồn giá trị i vào biến t (t = t + i)

print(t)            # In ra kết quả cuối cùng của t
```

```python
#W1A4
numbers = [1, 2, 3, 4]       # Tạo một danh sách gồm các số 1, 2, 3, 4

for x in numbers:            # Duyệt lần lượt từng phần tử trong danh sách (x lần lượt = 1, 2, 3, 4)
    if x % 2 == 0:           # Kiểm tra xem x có chia hết cho 2 hay không ( lỗi chưa lùi đầu dòng )
        print(x, "là số chẵn")   # Nếu chia hết cho 2, in ra "x là số chẵn" ( lỗi chưa lùi đầu dòng )
    else:
        print(x, "là số lẻ")     # Ngược lại, in ra "x là số lẻ" ( lỗi chưa lùi đầu dòng )
        
```

```python
#W1A5
animals = ["cat", "dog", "cat", "bird"]   # Tạo một danh sách gồm 4 phần tử: "cat", "dog", "cat", "bird"

count = 0                                 # Khởi tạo biến đếm count = 0

for a in animals:                         # Duyệt lần lượt từng phần tử trong danh sách animals
    count += 1                            # Mỗi lần duyệt 1 phần tử thì tăng biến count lên 1 ( lỗi chưa lùi đầu dòng )

print("Số phần tử trong danh sách là:", count)  
# Sau vòng lặp, count lưu tổng số phần tử trong danh sách
# In ra màn hình kết quả cuối cùng
```

```python
#W1A7
num = int(input("Nhập số: "))   # chuyển chuỗi nhập vào thành số nguyên
if num % 2 == 0:                # kiểm tra chia hết cho 2
    print("Số chẵn")            # in nếu là số chẵn
else:                           
    print("Số lẻ")              # in nếu là số lẻ
'''
input() trả về chuỗi (string)

Lỗi: num là chuỗi, không thể dùng toán tử %.

Cần chuyển sang số nguyên: num = int(num).

Sai toán tử trong if

Lỗi: if num % 2 = 0: → = là toán tử gán, không phải so sánh.

Sửa thành: if num % 2 == 0:.

Thiếu dấu : sau else

Lỗi: else phải viết thành else:.

Thiếu thụt lề (indentation)

Lỗi: print(...) trong if/else phải được thụt vào 4 khoảng trắng.

Nếu không, Python báo lỗi IndentationError.
'''
```

```python
#W1A8
for i in range(3):                    # Vòng lặp for chạy i = 0, 1, 2 (tổng cộng 3 lần)
    print("Al đang học lần", i + 1)   # Mỗi lần lặp in ra "Al đang học lần" kèm số thứ tự (i + 1)
print("Huấn luyện xong!")             # Sau khi vòng lặp kết thúc, in ra dòng kết thúc
'''Lỗi thụt lề (IndentationError)

Trong Python, khối lệnh bên trong vòng lặp for phải thụt vào (indent).

Ở đây dòng print("Al đang học lần", i + 1 ) chưa được thụt vào nên sẽ báo lỗi.

Cách viết thiếu rõ ràng

Dù không sai cú pháp, nhưng việc không có chú thích (comment) làm khó hiểu cho người mới.

(Ví dụ: không giải thích vì sao dùng i + 1 thay vì i).'''
```
