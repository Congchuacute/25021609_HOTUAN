```python
#b6
from math import isqrt
def snt(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

a, b = map(int, input().split())
d = 0
for i in range(a, b+1):
    if snt(i):
        d += 1
print(d)
```

```python
#b7
from math import isqrt
def snt(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
i = 1
ln = -10
while i*i <= n:
    if n % i == 0:
        j = n//i
        if snt(i):
            ln = max(ln,i)
        if snt(j):
            ln = max(ln,j)
    i+=1
print(ln)
```

```python
#b13
def ktr(n):
    d = 0
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            j = n//i
            if i != j:
                d = d + i + j
            else:
                d = d + i
    return d + 1
n, m = map(int, input().split())
if ktr(n) == m and  ktr(m) == n:
    print('true')
else:
    print('false')
    

```
