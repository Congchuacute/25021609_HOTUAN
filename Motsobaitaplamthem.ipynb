```python
#b57
n = []
m = []
t = []
while True:
    vl = input()
    n.append(vl)
    if vl == '':
        break
while True:
    vl = input()
    m.append(vl)
    if vl == '':
        break
c = len(n) + len(m)
i = 0
j = 0
while c > 0:
    if n[i] <= m[j] and i <= len(n) and j <= len(m):
        t.append(n[i])
        i += 1
        c -= 1
    if n[i] >= m[j] and j <= len(m) and j <= len(m):
        t.append(n[i])
        j += 1
        c -= 1
print(t)
```
